```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
```

```python
# data klasörüne ulaşmak için bir üst klasöre çıkıyoruz
df = pd.read_excel("../data/turkish_tweets.xlsx")

# Sütun isimlerini sadeleştir
df = df.rename(columns={"Tweet": "text", "Etiket": "label"})
df = df[["text", "label"]]
df.dropna(inplace=True)

df.head()
```

```python
le = LabelEncoder()
df["label_encoded"] = le.fit_transform(df["label"])
label_names = le.classes_
print("Etiketler:", label_names)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df["text"], df["label_encoded"], test_size=0.2, random_state=42
)
```

```python
import seaborn as sns

sns.countplot(x="label", data=df)
plt.title("Etiket Dağılımı")
plt.xticks(rotation=45)
plt.show()
```

```python
tfidf = TfidfVectorizer(max_features=5000, ngram_range=(1, 2))
X_train_vec = tfidf.fit_transform(X_train)
X_test_vec = tfidf.transform(X_test)

model = LogisticRegression(max_iter=1000)
model.fit(X_train_vec, y_train)
```

```python
y_pred = model.predict(X_test_vec)

print("Doğruluk:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred, target_names=label_names))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt="d", xticklabels=label_names, yticklabels=label_names, cmap="Blues")
plt.xlabel("Tahmin")
plt.ylabel("Gerçek")
plt.title("Confusion Matrix")
plt.show()
```
